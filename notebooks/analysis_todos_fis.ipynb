## carregando arquivos

```python
df = pd.read_csv(os.path.join(DATA_DIR, "todos-fis.csv"))
df.head()
```

```python
# Define a reusable function for running SQL queries
run_query = lambda query: sqldf(query, globals())
```

## Respondendo perguntas
- Quantidade de segmentos diferentes
- Quantos tipos de segmentos disponiveis
- Quantos setores diferentes
- Quantidade de fundos por segmento
- Quais tickers estão em cada segmento

```python
# qtde segmentos 
query_qtde_segmento = """
    SELECT COUNT(DISTINCT Segmento ) as qtde_segmentos
        FROM df
"""
result_qtde_segmento = run_query(query_qtde_segmento)
print(result_qtde_segmento)
```

```python
# quais os segmentos
query_seg_disp = """ 
    SELECT DISTINCT Segmento
        FROM df
"""
result_segmento_disp = run_query(query_seg_disp)
print(result_segmento_disp)
```

```python
# setores diferentes
query_setor = """ 
    SELECT DISTINCT Setor
        FROM df
"""
result_setor = run_query(query_setor)
print(result_setor)
```

```python
# qtde fundos por segmento
query_qtde_fundos_segmento = """ 
    SELECT COUNT(DISTINCT Segmento) as qtde_fundos_segmento,
            Setor
        FROM df
    GROUP BY Setor
"""
result_qtde_fundos_segmento = run_query(query_qtde_fundos_segmento)
print(result_qtde_fundos_segmento)
```

```python
# quais os tickers estao em cada segmento
query_ticker_segmento = """ 
   SELECT DISTINCT( Codigo ) as Codigo,
            Segmento    
        FROM df
    GROUP BY Segmento
    ORDER BY Segmento
"""
result_ticker_segmento = run_query(query_ticker_segmento)
print(result_ticker_segmento)
```
